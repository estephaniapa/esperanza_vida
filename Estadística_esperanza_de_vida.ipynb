## 6. Análisis estadístico inferencial

### 6.1 ¿La diferencia entre la esperanza media de vida de los paises desarrollados es mayor que en los paises en desarrollo? (Prueba de hipótesis)

<t align="justify">
Queremos verificar que la esperanza promedio en los países desarrollados y los países en vía de desarrollo es estadísticamente significativa. Esto puede abordarse a partir de la construcción de intervalos de confianza. Dado que la prueba "t" que se utiliza para esta comparación asume que las distribuciones son normales y tienen la misma dispersión, hemos decidido utilizar un enfoque no paramétrico, por esta razón utilizaremos la técnica de bootstrap para construir intervalos de confianza y comparar estos resultados para identificar diferencia significativa.

La técnica bootstrap es un método estadístico utilizado para aproximar la distribución de una estadística muestral, como la media, sin hacer suposiciones sobre la distribución subyacente de los datos. En lugar de obtener la distribución teórica exacta, la técnica bootstrap se basa en el muestreo con reemplazo de los datos disponibles para generar múltiples muestras "bootstrap" que son similares a la muestra original.

El proceso de bootstrap implica seguir los siguientes pasos:

Tomar una muestra de datos con reemplazo de la muestra original. Esto implica seleccionar aleatoriamente un punto de datos de la muestra original y devolverlo a la muestra antes de seleccionar el siguiente punto. Esto permite que un punto de datos se seleccione varias veces o no se seleccione en absoluto.

Calcular la estadística de interés (por ejemplo, la media) para cada una de las muestras bootstrap generadas en el paso anterior.

Repetir los pasos 1 y 2 un gran número de veces, generalmente miles de veces, para generar una distribución de la estadística muestral basada en las muestras bootstrap.

Una vez que se ha obtenido la distribución de la estadística muestral, se puede utilizar para construir intervalos de confianza. El intervalo de confianza proporciona un rango de valores dentro del cual es probable que se encuentre el verdadero valor del parámetro poblacional. Para construir un intervalo de confianza utilizando la técnica bootstrap, se seleccionan los percentiles específicos de la distribución de la estadística muestral. Por ejemplo, un intervalo de confianza del 95% se puede construir seleccionando el percentil 2.5 y el percentil 97.5 de la distribución bootstrap.

La técnica bootstrap es especialmente útil cuando no se cumplen los supuestos de normalidad o cuando la muestra es pequeña. Proporciona una forma robusta de aproximar la distribución de una estadística muestral y permite obtener intervalos de confianza más precisos sin depender de suposiciones teóricas.

```python
# Creamos nuevos dataframes para realizar gráficas
df_developed = df_life.loc[df_life['Status']=='Developed', 'Life expectancy '].dropna()
df_developing = df_life.loc[df_life['Status']=='Developing', 'Life expectancy '].dropna()

# Datos de las dos muestras
muestra1 = df_developed
muestra2 = df_developing
```

```python
fig = go.Figure()
# Graficamos
# Agregar los datos de los grupos al gráfico de caja
fig.add_trace(go.Box(y=muestra1, name='Desarrollados'))
fig.add_trace(go.Box(y=muestra2, name='En desarrollo'))

# Establecer el diseño del gráfico
fig.update_layout(
    title='Comparación de esperanza de vida',
    yaxis=dict(title='Esperanza de vida en años'),
    xaxis=dict(title='Nivel de desarrollo')
)

# Mostrar el gráfico
fig.show()
```

```python
# Generar múltiples muestras bootstrap
num_samples = 1000  # Número de muestras bootstrap a generar
```

```python
# Tamaños de muestra
tam_muestra = df_life['Status'].value_counts()
tam_muestra.head()
```

```python
# Definimos el dataframe donde guardamos los resultados
bootstrap_results = pd.DataFrame()

for dev in  {'Developed', 'Developing'}:
    bootstrap_result_cum = []
    for i in range(num_samples):
        # Generamos la muestra
        bootstrap_sample = resample(df_life.loc[df_life['Status']== dev], replace=True, n_samples= tam_muestra[dev])
        # Calculamos la media
        bootstrap_mean= bootstrap_sample['Life expectancy '].mean()
        # Guardamos resultados
        bootstrap_result_cum.append(bootstrap_mean)
    bootstrap_results[dev]=bootstrap_result_cum


print(bootstrap_results.head())
```

```python
#Gráfica comparación de las distribuciones muestrales simuladas mediante bootstrap
# Datos del histograma
x1 = bootstrap_results['Developing']
x2 = bootstrap_results['Developed']



# Crear la figura de histograma
fig = go.Figure()
fig.add_trace(go.Histogram(x=x1, name='Developing'))
fig.add_trace(go.Histogram(x=x2, name='Developed'))

# Configurar el diseño de la gráfica
fig.update_layout(
    title='Histograma de medias muestrales de esperanza de vida obtenidas mediante Bootstrap',
    xaxis_title='Esperanza de vida promedio en años',
    yaxis_title='Frecuencia',
    barmode='overlay',
    bargap=0.1
)
fig.update_traces(opacity=0.7)

# Mostrar la gráfica
fig.show()
```

A partir de la gráfica de arriba, vemos que de la técnica de bootstrap logramos aproximar las distribuciones de la media muestral de la esperanza de vida para distintos niveles de desarrollo. Es evidente, que al construir intervalos de confianza estos no se tocarán, por lo que podemos concluir que la diferencia en efecto es estadísticamente significativa. 

Procederemos a obtener dichos intervalos de confianza a partir de los percentiles de esta distribución aproximada, para obtener una estimación de los valores reales y de qué tan grande es la diferencia entre estas esperanzas de vida. Utilizaremos un nivel de confianza del 99%.

```python
NC=0.95

confidence_interval1 = [np.percentile(x1, (100-NC)/2), np.percentile(x1, 50+ (NC/2))]
confidence_interval2 = [np.percentile(x2, (100-NC)/2), np.percentile(x2, 50+ (NC/2))]
print(f'Intervalo de confianza para la esperanza de vida promedio de los paises en desarrollo: {confidence_interval1}')
print(f'Intervalo de confianza para la esperanza de vida promedio de los paises  desarrollados: {confidence_interval2}')
```

#### Conclusión
<t align="justify">

En conclusión, con un nivel de confianza del 99%, podemos afirmar que existe una diferencia significativa en la esperanza de vida promedio entre los países en desarrollo y los países desarrollados, siendo la esperanza de vida en los países desarrollados mayor en al menos 12 años en comparación con los países en desarrollo. Esta conclusión se basa en el análisis de datos y la realización de inferencia estadística no paramétrica, que indican que la diferencia observada no se debe simplemente al azar, sino que refleja una disparidad real en la esperanza de vida entre estos dos grupos de países. Estos hallazgos resaltan la importancia del desarrollo socioeconómico y la calidad de vida en la determinación de la esperanza de vida de una población, y subrayan la necesidad de implementar políticas y medidas para mejorar la salud y el bienestar en los países en desarrollo.

A continuación pasamos a obtener el primer modelo:

```python
# Seleccionamos las variables explicativas y target
X = df_estandarizado[['Schooling', 'Adult Mortality', ' BMI ', 'Developed', ' thinness  1-19 years', 'Polio']]  # Seleciona las columnas que serán tus variables independientes
y = df_estandarizado['Life expectancy ']  # Selecciona la columna que será tu variable dependiente

# Añadimos una constante
X = sm.add_constant(X)

# Creamos el modelo
model = sm.OLS(y, X)

# Ajustamos el modelo
results = model.fit()

print(results.summary())
```

<t align="justify">

Los resultados del modelo de regresión lineal múltiple indican lo siguiente:

- R-cuadrado (R-squared): El coeficiente de determinación R-squared es 0.737, lo que significa que aproximadamente el 73.7% de la variabilidad en la variable dependiente (esperanza de vida) puede ser explicada por las variables independientes incluidas en el modelo. Esto sugiere que el modelo tiene un buen ajuste y que las variables predictoras consideradas en el modelo son útiles para predecir la esperanza de vida.

- F-estadístico (F-statistic): El valor del F-estadístico es 765.4, y el valor p asociado es muy pequeño (prob (F-estadístico) = 0.00). Esto indica que el modelo en su conjunto es estadísticamente significativo, lo que significa que al menos una de las variables predictoras tiene un efecto significativo en la esperanza de vida.

- Coeficientes: Los coeficientes estimados para cada una de las variables predictoras muestran la relación entre cada variable y la esperanza de vida. Por ejemplo, el coeficiente para "Schooling" es 0.4541, lo que indica que un aumento de una unidad en el nivel educativo (Schooling) se asocia con un aumento de 0.4541 unidades en la esperanza de vida, manteniendo constantes las demás variables en el modelo.

- Estadísticos t y valores p: Para cada coeficiente, se proporciona un estadístico t y un valor p asociado. El estadístico t evalúa si el coeficiente estimado es significativamente diferente de cero. Los valores p indican la probabilidad de obtener un estadístico t tan extremo o más extremo si la hipótesis nula (el coeficiente es igual a cero) fuera verdadera. En este caso, todos los coeficientes tienen valores p muy pequeños (p < 0.05), lo que sugiere que todas las variables predictoras son estadísticamente significativas en relación con la esperanza de vida.

- En el caso de la constante (intercepto), el p-valor obtenido en la prueba individual t es alto (0.177), así que con un nivel de confianza del 95% no podemos rechazar la hipótesis nula de que este componente es significativo. Habría que plantear un nuevo modelo que no incluya este componente y evaluar si el modelo mejora.

- Errores estándar: Los errores estándar proporcionan una medida de la precisión de las estimaciones de los coeficientes. Un error estándar más pequeño indica una mayor precisión en la estimación del coeficiente.

Por lo tanto, los resultados del análisis de regresión lineal múltiple indican que las variables predictoras consideradas en el modelo (Schooling, Adult Mortality, BMI, Developed, thinness 1-19 years y Polio) tienen un impacto estadísticamente significativo en la esperanza de vida. Estos resultados proporcionan información valiosa para comprender y predecir los factores que influyen en la esperanza de vida en la población estudiada. 

Sin embargo, es importante tener en cuenta que el modelo asume varias hipótesis, entre ellas una relación lineal entre las variables, y que existen otras variables que pueden influir en la esperanza de vida y que no se han incluido en el modelo.

Además de los resultados anteriores, también tenemos información adicional para evaluar la calidad del modelo:

- Omnibus y Prob(Omnibus): El valor de Omnibus es 449.219 y el valor p asociado es 0.000. Esto indica que el modelo no cumple con la suposición de normalidad de los residuos. Es decir, existe evidencia de que los residuos del modelo no se distribuyen normalmente. El valor p pequeño sugiere que la desviación de la normalidad es estadísticamente significativa. Sin embargo, es importante tener en cuenta que el incumplimiento de la suposición de normalidad no invalida completamente el modelo, especialmente si el tamaño de la muestra es lo suficientemente grande.

- Durbin-Watson: El valor de Durbin-Watson es 0.839. Este estadístico se utiliza para evaluar la presencia de autocorrelación en los residuos. En este caso, el valor está cerca de 2, lo que sugiere que no hay evidencia sólida de autocorrelación en los residuos. Un valor entre 1.5 y 2.5 generalmente indica que no hay autocorrelación significativa.

- Jarque-Bera (JB) y Prob(JB): El valor de Jarque-Bera es 1783.576 y el valor p asociado es 0.00. Este estadístico se utiliza para evaluar la normalidad de los residuos del modelo. En este caso, el valor p es muy pequeño, lo que indica que los residuos no siguen una distribución normal. La asimetría negativa (skew) de -1.269 también respalda esta conclusión. Esto implica que los residuos tienen una distribución con colas más pesadas de lo esperado en una distribución normal.

En resumen, aunque el modelo de regresión lineal múltiple tiene un buen ajuste y las variables predictoras son significativas en relación con la esperanza de vida, se observa incumplimiento de algunas suposiciones clave del modelo, como la normalidad de los residuos. Esto sugiere que hay otros factores o variables no considerados en el modelo que pueden influir en la esperanza de vida y que podrían mejorar la precisión del modelo. Es importante tener precaución al interpretar y generalizar los resultados debido a estas limitaciones.


Procedemos a obtener un modelo de regresión lineal múltiple que no incluya al intercepto y veamos si nuestros resultados mejoran.


```python
# Seleccionamos las variables explicativas y target
X = df_estandarizado[['Schooling', 'Adult Mortality', ' BMI ', ' thinness  1-19 years', 'Polio']]  # Seleciona las columnas que serán tus variables independientes
y = df_estandarizado['Life expectancy ']  # Selecciona la columna que será tu variable dependiente


# Creamos el modelo
model = sm.OLS(y, X)

# Ajustamos el modelo
results = model.fit()

print(results.summary())
```

Notemos que el modelo mejoró en una centésima su coeficiente de determinación, sin embargo, los supuestos no se satisfacen. Veamos gráficamente con un análisis de residuales.

```python

y_pred = results.predict(X)


# Calcular los residuos
residuos = y - y_pred

# Graficar los residuales
plt.scatter(df_estandarizado['Life expectancy '], residuos)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Variable Independiente')
plt.ylabel('Residuales')
plt.title('Gráfico de dispersión de residuales')
plt.show()
```

A partir de esta gráfica podemos concluir que el supuesto de homocedasticidad no se satisface. No parece que la varianza de los residuales se mantenga constante. Ahora queremos ver si el supuesto de normalidad se satisface. Para esto, comenzaremos graficando los residuales así como también construyendo un QQ-plot para comparar el comportamiento con el de una distribución normal.

```python
# Graficar el histograma de residuales
plt.hist(residuos, bins=30)
plt.xlabel('Residuales')
plt.ylabel('Frecuencia')
plt.title('Histograma de residuales')
plt.show()
```

```python
from statsmodels.graphics.gofplots import qqplot

# Crear el QQ plot de los residuos
qqplot(residuos, line='s')

# Mostrar el gráfico
plt.show()
```

<t align="justify">

Al analizar las gráficas, podemos observar claramente que los residuos del modelo no siguen una distribución normal. El histograma muestra un sesgo hacia la izquierda, lo que indica una concentración de residuos con valores más bajos de lo esperado. Esto puede indicar la presencia de factores no considerados en el modelo que están influyendo en los resultados.

Además, al examinar el QQ-plot, podemos confirmar que los residuos no se ajustan a una línea recta. En lugar de seguir una distribución normal, los puntos del QQ-plot se desvían de la línea diagonal. Esto refuerza la evidencia de que los residuos no tienen un comportamiento normal y sugiere la presencia de otras fuentes de variabilidad que no están siendo capturadas por el modelo.

Es importante tener en cuenta que el incumplimiento del supuesto de normalidad no invalida completamente el modelo, especialmente si el tamaño de la muestra es lo suficientemente grande. Sin embargo, debemos ser cautelosos al interpretar los resultados y considerar la posibilidad de incluir más variables explicativas o utilizar técnicas alternativas que puedan abordar la falta de normalidad en los residuos, como modelos no lineales o transformaciones de las variables.
